```python
import matplotlib.pyplot as plt

x=['S','A','R','O','J']
y=[2,4,6,8,10]
plt.plot(x,y,marker="*")
plt.xlabel("Name letters")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'Score': [85, 90, 95]}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Basic operations
print("Mean age:", df['Age'].mean())
print("Data sorted by Score:\n", df.sort_values(by='Score', ascending=False))
```

```python

import matplotlib.pyplot as plt

# Sales data
products = ['Pen', 'pencils','stapler', 'copy', 'Book']
sales = [120, 150, 90, 150, 170]

# Creating the bar chart
plt.bar(products, sales, color='skyblue')
plt.title('Stationary Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```
